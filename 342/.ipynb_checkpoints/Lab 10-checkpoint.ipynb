```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy import optimize
```

#### 1)

```python
#Method approximates endpoints, and returns them in a list in the form of
#[Lower bound, Upper bound]
def method1(x):
    #Case for when x is negative; empty list returned
    if x < 0:
        return []
    #Case for x - sqrt(x) < 0
    elif x - np.sqrt(x) < 0:
        lb = 0
        ub = x  + np.sqrt(x)
        return [lb,ub]
    #Everything else
    else:
        lb = x - np.sqrt(x)
        ub = x + np.sqrt(x)
        return [lb,ub]
```

#### 2)

By squaring the boundary points in the approximate method and solving for theta,

i.e. $\theta \pm \sqrt\theta = x$.

This gives $\theta = (x \mp \theta)^2$

From this, two resulting Neyman equations were found to be 

$\theta = x + (\frac{1 - \sqrt{1+4x}}{2})$ for the lower endpoint

and

$\theta = x + (\frac{1 + \sqrt{1+4x}}{2})$ for the upper endpoint, by using the quadratic formula and plugging in the corresponding coefficients.

```python
theo_theta = np.linspace(0,2,500)
x_lower = theo_theta - np.sqrt(theo_theta)
x_upper = theo_theta + np.sqrt(theo_theta)
x_mid = theo_theta
```

```python
#Used to find the index in a list that has a specific value
def findIndex(li,val):
    count = 0
    indexes = []
    eps = 1.8e-3
    for i in li:
        if abs(i - val) <= eps:
            indexes.append(count)
        count += 1
    return indexes
```

```python
obs = [-0.2,0.,0.5]
line1 = findIndex(x_lower, obs[1])
print(line1)
#Since index 249 and 250 are evenly spaced around 0, the range is 0 ---> 249
th1 = theo_theta[:250]
```

```python
line2 = findIndex(x_lower, obs[0])
print(line2)
#Range is chosen to be from 19 ---> 131
th2 = theo_theta[19:132]
```

```python
plt.figure(figsize = (12,5))
plt.plot(x_lower,theo_theta, label = 'high')
#Functions for plotting verticle lines within a range
plt.vlines(x=obs[1], ymin = th1[0], ymax = th1[-1], label = "obs = 0.")
plt.vlines(x=obs[0], ymin = th2[0], ymax = th2[-1], label = "obs = -0.2", color = 'r')
plt.vlines(x=obs[2], ymin = theo_theta[34], ymax = theo_theta[465], label = "obs = 0.5", color = 'm')
plt.plot(theo_theta, x_mid, label = 'expectation')
plt.plot(x_upper,theo_theta, label = 'low')
plt.xlabel("x")
plt.ylabel(r'$\theta$')
plt.title("Neyman Confidence Belt")
plt.legend()
```

#### 3)

```python
x = [-0.2,0.,0.5,1.,2.]
#Finding the approximate intervals
for i in x:
    approx = method1(i)
    print("For observation",i, "the interval is:", approx)
```

```python
#Using Neyman functions found above
func1 = lambda x: x + (1 + np.sqrt(1 + 4*x)) * 0.5
func2 = lambda x: x + (1 - np.sqrt(1 + 4*x)) * 0.5
for i in x:
    sol_1 = func1(i)
    sol_2 = func2(i)
    li = [sol_2, sol_1]
    print("For observation", i, "the interval using the Neyman equations is", li)
```

#### 4)

```python
func1 = lambda th,x: -th + x + (1 + np.sqrt(1 + 4*x)) * 0.5
func2 = lambda th,x: -th + x + (1 - np.sqrt(1 + 4*x)) * 0.5
```

```python
theta = [0.1,0.2,1.]
for i in theta:
    count_1 = 0
    count_2 = 0
    #Solving the boundary equations for each theta
    sol_1 = optimize.fsolve(func1,x0 = 0, args = (i))
    sol_2 = optimize.fsolve(func2,x0 = 0, args = (i))
    for j in range(100000):
        test = np.random.normal(i,np.sqrt(i),1)
        #Approximating the bounds using method 1
        approx = method1(i)
        if test >= approx[0] and test <= approx[1]:
            count_1 += 1
        #Checking to see if the Neyman solutions are within the bounds
        if test >= sol_2 and test <= sol_1:
            count_2 += 1
    print("The fraction of approximate intervals that contain the true value", i, "is", count_1/100000)
    print("The fraction of Neyman intervals that contain the true value", i, "is", count_2/100000)
```
