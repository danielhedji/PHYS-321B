```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

### 6d)

```python
#Function describing the x component of the electric field
def E_x(E,x_r,y_r,z_r):
    term1 = y_r/((x_r**2 + y_r**2) ** 0.5)
    term2 = (x_r-1)/(((x_r-1)**2 + (y_r - 1)**2 + z_r**2) ** 1.5)
    return term1 * term2
```

```python
#Function describing the y component of the electric field
def E_y(E,x_r,y_r,z_r):
    term1 = y_r/((x_r**2 + y_r**2) ** 0.5)
    term2 = (y_r-1)/((x_r-1)**2 + (y_r - 1)**2 + z_r**2) ** 1.5
    return term1 * term2
```

```python
#Function meant for handling the integral
def integrand(E,x_r,y_r,z_r):
    return 1/(4*np.pi) * quad(E, -1, 1, args=(x_r, y_r, z_r))[0]
```

```python
#constants given
y_r = 0.5
x_r = np.linspace(-0.5,0.5,1000)
z_r = 0
```

```python
#Evaluating the integrals
int_x = np.vectorize(integrand)(E_x,x_r,y_r,z_r)
int_y = np.vectorize(integrand)(E_y,x_r,y_r,z_r)
```

```python
#Plotting for E_x
plt.plot(x_r,int_x, color = 'g')
plt.xlabel('$x/R$', fontsize=16)
plt.ylabel(r'$\frac{4 \pi \epsilon_0 R^2}{\lambda_0}\vec{E_x}$', fontsize=16)
plt.title("The electric field's x-component as a function of $x/R$")
```

```python
#Plotting E_y
plt.plot(x_r, int_y, color = 'r')
plt.xlabel('$x/R$', fontsize=16)
plt.ylabel(r'$\frac{4 \pi \epsilon_0 R^2}{\lambda_0}\vec{E_y}$', fontsize=16)
plt.title("The electric field's y-component as a function of $x/R$")
```
