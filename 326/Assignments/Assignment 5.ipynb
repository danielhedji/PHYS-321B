```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
from scipy.special import legendre
```

```python
#Code for the integrand of the multipole expansion
def integrand(phi,n,N,x_r,y_r,z_r):
    return legendre(abs(n))((x_r * np.cos(phi) + y_r * np.sin(phi)) / (np.sqrt(x_r**2 + y_r**2 + z_r**2))) * np.sin(N * phi)
```

```python
#The entire function V (dimensionless)
def func(n,N,x_r,y_r,z_r,r_r):
    integral, error = quad(integrand,0,2 * np.pi, args = (n,N,x_r,y_r,z_r))
    if np.abs(error) > np.abs(integral):
        integral = 0
    return integral * 1/(r_r**(n+1))
```

```python
#Making various n and phi values
n = np.linspace(0,9,10)
phi = np.linspace(0,2*np.pi,10)
```

### 1b)

```python
#Corresponding V values for each r
#Note that the components of each r were input manually, with the last argument being the magnitude of each r
V_11 = np.vectorize(func)(n,1,6,10,12,16.73)
V_21 = np.vectorize(func)(n,1,5,9,7,12.45)
V_31 = np.vectorize(func)(n,1,7,4,11,13.64)
```

```python
#Constants for bar plot aesthetic
barwidth = 0.2
ls1 = n - barwidth/2
ls2 = n + barwidth/2
```

```python
#Plotting the found data (absolute value)
plt.figure(figsize=(12,5))
plt.bar(ls1, np.abs(V_11),width = barwidth, label = "r1")
plt.bar(n, np.abs(V_21),width = barwidth, label = "r2")
plt.bar(ls2, np.abs(V_31),width = barwidth, label = "r3")
plt.semilogy()
plt.xlabel('$n$')
plt.ylabel(r'$|\frac{4 \pi\epsilon_0 V}{\lambda_0}|$')
plt.minorticks_off()
plt.xticks(n)
plt.title('Multipole expansion for N=1')
plt.legend()
plt.grid()
plt.show()
```
