```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

### 5b)

```python
#Formula inside the series for W
def series_W(N):
    tot = 0 
    num = 1
    while num != N:
        term = 1 / np.sin(np.pi * num / N)
        tot += term
        num += 1
    return tot / N
```

```python
x_data = np.arange(2,25)
y_data = y_data = np.array([series_W(N) for N in x_data])
```

```python
plt.plot(x_data,y_data, label = "Work done")
plt.xlabel('$N$', fontsize=14)
plt.ylabel(r'$\left( \frac{16 \pi \epsilon_0 R}{Q^2} \right)W$', fontsize=14)
plt.title("Work required to assemble a ring of charge as a function of increasing point charges")
plt.grid()
plt.legend()
plt.show()
```

```python
#Function used in the curve fit
def f(N,a,b,c):
    return a*(np.log(N)**b) + c
```

```python
#Applying curve fit
fit_results = curve_fit(f,x_data,y_data)
```

```python
#Seeing what is returned
fit_results
```

```python
#Extracting the approximations for the a, b, and c constants
a_fit = fit_results[0][0]
b_fit = fit_results[0][1]
c_fit = fit_results[0][2]
```

```python
a_fit
```

```python
b_fit
```

```python
c_fit
```

```python
plt.plot(x_data, y_data, label='Work done')
plt.plot(x_data, f(x_data, a_fit,b_fit,c_fit), label='Best Fit', ls='--')
plt.xlabel('$N$', fontsize=14)
plt.ylabel(r'$\left( \frac{16 \pi \epsilon_0 R}{Q^2} \right)W$', fontsize=14)
plt.title("Work as a function of N")
plt.grid()
plt.legend()
plt.show()
```
