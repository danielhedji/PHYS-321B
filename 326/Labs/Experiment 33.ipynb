```python
def func(z,a,b):
    return (a / z **3) + b
```

```python
z = []
for i in distance:
    z.append(func(i,1,0))
z = np.array(z)
distance = np.array(distance)
```

```python
popt, pcov = curve_fit(func,distance,z, p0 = [1,0])
```
