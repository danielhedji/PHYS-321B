```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Inputting the raw data from LoggerPro
data = np.loadtxt(r"data.txt")
data
```

### Format of the above data arrays
*Note that each number represents column number*

0. Time (seconds)
1. x position of puck 1 (meters)
2. y position of puck 1 (meters)
3. x velocity of puck 1 (m/s)
4. y velocity of puck 1 (m/s)
5. x position of puck 2 (meters)
6. y position of puck 2 (meters)
7. x velocity of puck 2 (m/s)
8. y velocity of puck 2 (m/s)

```python
#Constants
m_puck = 0.278 #+/- 5e-3. Unit of kg
d_puck = 92e-3 #+/- 0.05e-3. Unit of m
```

### Center of mass

```python
#Inputting in the x and y columns for both pucks
#The unit of each data point is in meters
x1 = data[:,1]
y1 = data[:,2]
x2 = data[:,5]
y2 = data[:,6]
```

```python
#Finding the center of mass
x_avg = []
y_avg = []
#Iterating through each list at once
for i,j,k,l in zip(x1,x2,y1,y2):
    #Applying formula
    x_avg.append((i+j)/2)
    y_avg.append((k+l)/2)
```

```python
#Plotting
plt.figure(figsize = (12,5))
plt.plot(x_avg,y_avg, label = "$ \\frac{y_{avg}}{x_{avg}}$")
plt.grid()
plt.xlabel("X position of the centre of mass (m)")
plt.ylabel("Y position of the centre of mass (m)")
plt.title("Average y against average x of the center of mass")
plt.legend()
```

```python
time = data[:,0] #Extracting the time element (in s)
#Plotting results
plt.figure(figsize = (12,5))
plt.plot(time,x_avg, label = "$x_{avg}$")
plt.plot(time,y_avg, label = "$y_{avg}$")
plt.grid()
plt.xlabel("Time (s)")
plt.ylabel("Average of a coordinate (m)")
plt.title("Average coordinates of the center of mass as a function of time")
plt.legend()
```

### Linear momentum

```python
vx1 = data[:,3]
vy1 = data[:,4]
vx2 = data[:,7]
vy2 = data[:,8]
```

```python
mom_x1 = []
mom_x2 = []
for i,j in zip(vx1,vx2):
    mom_x1.append(m_puck * i)
    mom_x2.append(m_puck * j)
```

```python
plt.figure(figsize = (12,5))
plt.plot(time, mom_x1,".", label = "puck 1")
plt.plot(time, mom_x2,".", label = "puck 2")
plt.grid()
plt.xlabel("Time (s)")
plt.ylabel("x-component of linear momentum (kgm/s)")
plt.title("x-component of linear momentum as a function of time")
plt.legend()
```

```python
mom_y1 = []
mom_y2 = []
for i,j in zip(vy1,vy2):
    mom_y1.append(m_puck * i)
    mom_y2.append(m_puck * j)
```

```python
plt.figure(figsize = (12,5))
plt.plot(time, mom_y1, ".", label = "puck 1")
plt.plot(time, mom_y2, ".",label = "puck 2")
plt.grid()
plt.xlabel("Time (s)")
plt.ylabel("y-component of linear momentum (kgm/s)")
plt.title("y-component of linear momentum as a function of time")
plt.legend()
```

### Angular momentum

```python
#Solving for r for each frame
r = [] #this will have units m
#Iterating through each list at once
for i,j,k,l in zip(x1,x2,y1,y2):
    termx = (j-i)**2
    termy = (l-k)**2
    r.append(np.sqrt(termx + termy))
```

```python
#Finding the angle theta for each frame
th = []
for i,j,t in zip(x1,x2,r):
    term = (j-i) / t
    th.append(np.arccos(term))
```

```python
#Computing the changes in time for each frame
delt_t = []
for i in range(1,len(time)):
    delt_t.append(time[i] - time[i-1])
```

```python
#Finding w by the change of theta per frame
om = []
for i in range(len(th)-1):
    if i == 0:
        change = th[i]
        om.append(change)
    else:
        change = th[i] - th[i-1]
        om.append(change/(delt_t[i]))
```

```python
L1 = [] 
#Applying eqn. 15.1
for i in range(len(r)-1):
    L = om[i] * ((0.5 * m_puck * r[i] ** 2) + (m_puck * (d_puck/2) **2))
    L1.append(L)
```

```python
#Finding the mean
mu = np.mean(L1)
mu
```

```python
#Finding the percent differences
differences = []
for i in L1:
    dif = (i - mu) * 100 / mu
    differences.append(dif)
print("The percent differences of the angular momentum will be within the range", min(differences), "and", max(differences))
```

```python
#Since the values are vary significantly, the median of the range is taken
avg = np.median(differences)
print("On average, there will be a percent difference of", round(avg,3))
```

```python
plt.figure(figsize = (12,5))
plt.plot(time[:-1],L1, "-", label = "L")
plt.grid()
plt.xlabel("Time (s)")
plt.ylabel("Angular momentum $ (\\frac{kgm^2}{s})$")
plt.title("Angular momentum as a function of time")
plt.legend()
```

```python
#Inversing the omega values
om_inv = []
for i in om:
    om_inv.append(1/i)
#Squaring the r values
r_sq = []
for j in r:
    r_sq.append(j**2)
```

```python
plt.figure(figsize = (12,5))
plt.plot(r_sq[:157],om_inv[:157], "b.")
plt.grid()
plt.xlabel("$r^2$ $(m^2)$")
plt.ylabel("$ \\frac{1}{\omega}$")
plt.title("$\\frac{1}{\omega}$ against $r^2$")
```

```python
#Applying a linear fit to the data
poly = np.polyfit(r_sq[:157],om_inv[:157], deg = 1)
print("The magnitude of the angular momentum has a slope of", poly[0], "and an intercept of", poly[1])
```

```python
#Creating a range for r values
r_p = np.linspace(0.025,0.15,156)
#Follows the linear line general formula y=mx+b
#Applying the parameters to the equation above
w_p = poly[0] * r_p + poly[1]
#Plotting results
plt.figure(figsize = (12,5))
plt.plot(r_sq[:157],om_inv[:157], "y.", label = "data")
plt.plot(r_p, w_p, label = "fit")
plt.grid()
plt.xlabel("$r^2$ $(m^2)$")
plt.ylabel("$ \\frac{1}{\omega}$")
plt.title("$\\frac{1}{\omega}$ against $r^2$")
plt.legend()
```
