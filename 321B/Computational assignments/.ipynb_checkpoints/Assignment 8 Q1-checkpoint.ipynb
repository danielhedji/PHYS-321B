The 3 equations of motion come from the lecture notes and are as follows

(1) $I_1 \dot{\omega_1} = -\omega_2\omega_3(I_3-I_2)$

(2) $I_2 \dot{\omega_2} = -\omega_1\omega_3(I_1-I_3)$

(3) $I_3 \dot{\omega_3} = -\omega_1\omega_2(I_2-I_1)$

To find the angular momentum over time, the system needs to be integrated with respect to time. Since $I_1, I_2,$ and $I_3$ are all unknown, the process used to find the tensor elements in assignment 7, question 1a) were used to find these elements. Note that these are the diagonal elements in the moment of inertia tensor, $I^{11}, I^{22},$ and $I^{33}$

Analytically, these were found to be $\frac{12}{3}, \frac{8}{3},$ and $\frac{8}{3}$, respectively.

```python
#Initial conditions, all in 1/s
wx_1, wx_2, wx_3 = 0, 20, 1
wy_1, wy_2, wy_3 = 1, 0, 20
wz_1, wz_2, wz_3 = 20, 1, 0
```

```python
def L(w_x,w_y,w_z):
    #Analytically found I values
    I_1 = 4.
    I_2 = 8/3
    I_3 = 8/3
    
    #Timestep
    ts = 1e-4
    t = 0
    #Defining constants
    w_1, w_2, w_3 = w_x, w_y, w_z
    L1, L2, L3 = [], [], []
    
    while t <= 3:
        #Applying eom
        w1_d = -w_2 * w_3 * (I_3 - I_2) / I_1
        w2_d = -w_1 * w_3 * (I_1 - I_3) / I_2
        w3_d = -w_1 * w_2 * (I_2 - I_1) / I_3
        #Forward integrating the omega values
        w_1 = w1_d * ts + w_1
        w_2 = w2_d * ts + w_2
        w_3 = w3_d * ts + w_3
        #Finding the overall scalar angular momentum
        L_val = np.sqrt((w_1 * I_1) ** 2 + (w_2 * I_2) ** 2 + (w_3 * I_3) ** 2)
        #Organizing the values
        L1.append(w_1 * I_1)
        L2.append(w_2 * I_2)
        L3.append(w_3 * I_3)
        t += ts
        
    return L1,L2,L3
```

```python
t = np.linspace(0,3,30000)
L1, L2, L3 = L(wx_1,wy_1,wz_1) 
```

```python
#Plotting
plt.figure(figsize = (12,5))
plt.plot(t, L1, label = "$L_1$")
plt.plot(t, L2, label = "$L_2$")
plt.plot(t, L3, label = "$L_3$")
plt.grid()
plt.xlabel("Time (s)")
plt.ylabel("L ($\\frac{kgm^2}{s}$)")
plt.title("Angular momentum values over time")
plt.legend()
```
