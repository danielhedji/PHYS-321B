```python
import numpy as np
import matplotlib.pyplot as plt
```

### 2)

```python
#Values of the moments of inertia found from the lab
I_1e = 4.01e4 #in cm^2/g
I_2e = 3.89e4 #in cm^2/g
I_3e = I_2e 
```

```python
#Case 1
omega = [0.5,1.,3.]
th_d1 = 0.
phi_d1 = 0.2
```

```python
def traj(th_d,phi_d, I_1, I_2, I_3, om):
    #Timestep
    ts = 1e-4
    t = 0 
    #Psi constant btw
    th, phi = 0., 0.
    th_s, phi_s = [], []
    while t  < 10:
        #Applying the lab textbook equations 18.6 an 18.10
        #and forward integrating to find the evolution
        phi_d = (I_1 * om/ I_3) * th
        phi = phi_d * ts
        
        th_dd = (1/I_3) - (I_1*om/I_3)**2 * th
        th_d = th_dd * ts + th_d
        th = th + th_d * ts + 0.5 * th_dd * ts ** 2
        
        th_s.append(th)
        phi_s.append(phi)
        
        t += ts
        
    return th_s, phi_s
```

```python
t = np.linspace(0,10,100001)
#Finding the theta and phi values by passing parameters into the function
th, phi = traj(th_d1,phi_d1,I_1e,I_2e,I_3e,omega[0])
```

```python
#Plotting
plt.figure(figsize = (12,5))
plt.plot(t,th)
plt.grid()
plt.xlabel("Time (s)")
plt.ylabel("$\\theta$ (rad)")
plt.title("$\\theta$ values over time for $\omega$, $\\dot{\\theta}$,  $\\dot{\\phi}$ =" + str(omega[0]) + "," + str(th_d1)
          + "," + str(phi_d1))
```

```python
plt.figure(figsize = (12,5))
plt.plot(t,phi)
plt.grid()
plt.xlabel("Time (s)")
plt.ylabel("$\\phi$ (rad)")
plt.title("$\\phi$ values over time for $\omega$, $\\dot{\\theta}$,  $\\dot{\\phi}$ =" + str(omega[0]) + "," + str(th_d1)
          + "," + str(phi_d1))
```
