```python
import numpy as np
import matplotlib.pyplot as plt
```

For this problem, the 3 equations of motion for $x_1, x_2, x_3$ are 

$m_1\ddot{x_1} = -k_1x_1-k_2(x_1-x_2)$,

$m_2\ddot{x_2} = k_2(x_1-x_2) - k_3(x_2-x_3)$,

$m_3\ddot{x_3} = k_3(x_2-x_3)-k_4x_3$,

respectively.

```python
#Taking the data from the csv file given
text = np.loadtxt(r'ML3-Glider-M09.txt')
text
```

```python
#Initial conditions
m1,m2,m3 = 320,160,160 #in g
k1,k2,k3,k4 = 1.6, 0.8, 1.6, 1.6 #in N/m
x11,x21,x31 = -6.8e-2,-2.6e-2, 1.5e-2 #in cm
x12,x22,x32 = 0.,2.4e-2,-2.5e-2 #in cm
x13,x23,x33 = -8.8e-2,0.2e-2,0. #in cm
```

```python
#Converting to the same base (kg)
m1 = m1 / 1000
m2 = m2 / 1000
m3 = m3 / 1000
```

```python
def forward_int(xo1,xo2,xo3,m1,m2,m3,k1,k2,k3,k4,t):
    #Lists to store acceleration values
    dd1, dd2, dd3 = [], [], [] 
    #Initial conditions
    x_d1, x_d2, x_d3 = 0., 0., 0.
    ts = 0.02 #in s
    x1 = xo1 #in sm
    x2 = xo2 #in m
    x3 = xo3 #in m
    t_tot = 0 #in s
    
    while t_tot <= t:
        #Applying eom above to find accelerations
        x_dd1 = (1/m1) * (-k1*x1 - k2*(x1-x2))
        dd1.append(x_dd1)
        x_dd2 = (1/m2) * (k2*(x1-x2) - k3*(x2-x3))
        dd2.append(x_dd2)
        x_dd3 = (1/m3) * (-k4*x3 + k3*(x2-x3))
        dd3.append(x_dd3)
        #Position of masses
        x1 = 0.5*x_dd1*ts**2 + x_d1*ts + x1
        x2 = 0.5*x_dd2*ts**2 + x_d2*ts + x2
        x3 = 0.5*x_dd3*ts**2 + x_d3*ts + x3
        #Forward integrating for velocity of masses
        x_d1 = x_dd1 * ts + x_d1
        x_d2 = x_dd2 * ts + x_d2
        x_d3 = x_dd3 * ts + x_d3
        #Incrementing time
        t_tot += ts
        
    return dd1,dd2,dd3
```

```python
#Extracting the acceleration data for 1st trial
t1 = text[:,0] #Time column
a11 = text[:,1] #M1 acceleration
a21 = text[:,2] #M2 acceleration
a31 = text[:,3] #M3 acceleration
```

```python
a12 = text[:,5] #M1 acceleration
a22 = text[:,6] #M2 acceleration
a32 = text[:,7] #M3 acceleration
```

```python
a13 = text[:,9] #M1 acceleration
a23 = text[:,10] #M2 acceleration
a33 = text[:,11] #M3 acceleration
```

```python
#Theoretical acceleration data
a1_t = forward_int(x11,x21,x31,m1,m2,m3,k1,k2,k3,k4,t1[-1])
```
