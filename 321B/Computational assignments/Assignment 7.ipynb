### 1.

For this problem, the 3 equations of motion for $x_1, x_2, x_3$ are 

$m_1\ddot{x_1} = -k_1x_1-k_2(x_1-x_2)$,

$m_2\ddot{x_2} = k_2(x_1-x_2) - k_3(x_2-x_3)$,

$m_3\ddot{x_3} = k_3(x_2-x_3)-k_4x_3$,

respectively.

```python
#Taking the data from the csv file given
text = np.loadtxt(r'ML3-Glider-M09.txt')
text
```

```python
#Initial conditions
m1,m2,m3 = 320,160,160 #in g
k1,k2,k3,k4 = 1.6, 0.8, 1.6, 1.6 #in N/m
x11,x21,x31 = -6.8,-2.6, 1.5 #in cm
x12,x22,x32 = 0.,2.4,-2.5 #in cm
x13,x23,x33 = -8.8,0.2,0. #in cm
```

```python
def forward_int(xo1,xo2,xo3,m1,m2,m3,k1,k2,k3,k4,t):
    #Lists to store acceleration values
    dd1, dd2, dd3 = [], [], [] 
    #Initial conditions
    x_d1, x_d2, x_d3 = 0., 0., 0.
    ts = 0.02 #in s
    x1 = xo1 #in sm
    x2 = xo2 #in m
    x3 = xo3 #in m
    t_tot = 0 #in s
    
    while t_tot <= t:
        #Applying eom above to find accelerations
        x_dd1 = (1/m1) * (-k1*x1 - k2*(x1-x2))
        dd1.append(x_dd1)
        x_dd2 = (1/m2) * (k2*(x1-x2) - k3*(x2-x3))
        dd2.append(x_dd2)
        x_dd3 = (1/m3) * (-k4*x3 + k3*(x2-x3))
        dd3.append(x_dd3)
        #Position of masses
        x1 = 0.5*x_dd1*ts**2 + x_d1*ts + x1
        x2 = 0.5*x_dd2*ts**2 + x_d2*ts + x2
        x3 = 0.5*x_dd3*ts**2 + x_d3*ts + x3
        #Forward integrating for velocity of masses
        x_d1 = x_dd1 * ts + x_d1
        x_d2 = x_dd2 * ts + x_d2
        x_d3 = x_dd3 * ts + x_d3
        #Incrementing time
        t_tot += ts
        
    return dd1,dd2,dd3
```

```python
#Extracting the acceleration data for 1st trial
t1 = text[:,0] #Time column
a11 = text[:,1] #M1 acceleration
a21 = text[:,2] #M2 acceleration
a31 = text[:,3] #M3 acceleration
```

### 2.

For this given problem, the $-\omega^2T + V$ matrix is $\begin{pmatrix} -m_1\omega^2-(k_1+k_2) & k_2 & 0 \\ k_2 & -m_2\omega^2-(k_2+k_3) & k_3 \\ 0 & k_3 & -m_3\omega^2-(k_3+k_4) \end{pmatrix}$. To find the $\omega^2$ values, the determinant of the above matrix needs to be $0$.

Analytically solving the determinant, gives 

$[-m_1\omega^2-(k_1+k_2)][(-m_2\omega^2-(k_2+k_3))(-m_3\omega^2-(k_3+k_4))-k_3^2]-k_2[(k_2(-m_3\omega^2-(k_3+k_4))] = 0$

```python
#Omega values
w = np.linspace(0.05,0.2,10000)
```

```python
deter = []

#Quantities defined to simplify the calculation
alpha = m2*m3
beta = (m2+m3) * (k3+k4)
gamma = (k2+k3) * (k3+k4)
#Coefficient values
term1 = m1*alpha
term2 = m1*beta + alpha*(k1+k2)
term3 = m1*gamma + beta*(k1+k2)
term4 = gamma*(k1+k2)
for i in w:
    deter.append(-term1*i**6+term2*i**4-term3*i**2+term4)
```

```python
plt.plot(w,deter)
plt.grid()
plt.xlabel("$\omega$ (Hz)")
plt.ylabel("Determinant value")
plt.title("Determinant values as a function of $\omega^2$")
```

```python
#Freq. array with arbitrarily chosen frequencies
freqs = np.linspace(0.1,6,300)
```

```python
def fourier_cos(f,times,accel):
    #Timestep found within the data
    ts = 0.02
    FC = [] #Array for the fourier cosine transform values
    for i in range(len(f)):
        fc = np.cos(2*np.pi*f[i]*times) #Cosine term in the fourier transform
        conv = fc * accel #Value of the integrand
        integral = conv.sum() * ts #Integrating for the given timestep
        FC.append(integral)
    return FC

def fourier_sin(f,times,accel):
    #Timestep found within the data
    ts = 0.02
    FS = [] #Array for fourier sine transform values
    for i in range(len(f)):
        fs = np.sin(2*np.pi*f[i]*times) #Sine term in the fourier transform
        conv = fs * accel #Finding the value of the integrand
        integral = conv.sum() * ts #Integrating for the given timestep
        FS.append(integral)
    return FS
```

```python
#Applying the fourier transforms to each of the masses for the first trial
FC_11 = fourier_cos(freqs,t1,a11)
FS_11 = fourier_sin(freqs,t1,a11)
FC_21 = fourier_cos(freqs,t1,a21)
FS_21 = fourier_sin(freqs,t1,a21)
FC_31 = fourier_cos(freqs,t1,a31)
FS_31 = fourier_sin(freqs,t1,a31)
```

```python
plt.plot(freqs, FC_11,label = "Fourier cos")
plt.plot(freqs, FS_11, label = "Fourier sin")
plt.plot(freqs, np.sqrt(np.array(FC_11)**2 + np.array(FS_11)**2), label = "Fourier mag.")
plt.legend()
plt.grid()
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude ($\\frac{cm^2}{s}$)")
plt.title("Fourier analysis as a function of frequency for mass 1, trial 1")
```
