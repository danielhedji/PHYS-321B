```python
import numpy as np
import matplotlib.pyplot as plt
import math as ma
```

The Routhian for this problem is $\ddot{r} - \frac{p_\theta^2}{m^2r^3} + \frac{1}{mr^2} = 0$. This also gives an angular momentum of $p_\theta = mr^2\dot{\theta}$

```python
def scatter(m,xo,vo,s):
    ts = 1e-4
    eps = 1e-4
    #Applying the above eqns/I.C.
    r = (xo**2 + s**2)**0.5
    r_d = vo
    p_th = m*s*vo
    th = np.sin(s/r)
    #checking to see if r is approx x0 and v is approx v0
    while (1.5*r_d - abs(vo)) > eps and (1.5*r - xo) > eps:
        r_dd = p_th**2/(m**2 * r**3) - 1/(m*r**2)
        th_d = p_th/(m*r**2)
        #Forward integrating
        r_d = r_dd * ts + r_d
        r = r_d * ts + r
        th = th_d * ts + th
        #Updating p_theta
        p_th = m*s*r_d
        
    return r,th
```

```python
#Constants
s = np.linspace(1,25,1000)
m = 1
xo = 1000
vo = -3
```

```python
#Finding the r and thetas for various s values
r = []
th1 = []
for i in s:
    a,b = scatter(m,xo,vo,i)
    r.append(a)
    th1.append(b)
```

```python
plt.plot(s,th1)
plt.xlabel("s")
plt.ylabel("Theta")
plt.title("Theta as a function of s")
```

```python
#Applying definition of derivative
diff_t = []
diff_r = []
for i in range(1,len(th1)):
    diff_t.append(th1[i]-th1[i-1])
    diff_r.append(r[i]-r[i-1])
```

```python
plt.plot(diff_t,diff_r)
plt.xlabel("ds")
plt.ylabel("d$\\theta$")
plt.title("d$\\theta$ as a function of ds")
```
