In this problem, I treated my solution process in a similar manner to what I used in the eighth computational assignment. First I needed a way to get each of the respective $\omega$ values, so I applied the definition of angular momentum in matrix form. Next, I calculated the moment of inertia tensor (using the diagonal values only), and then I had all of the components needed to continue. For increasing time, I found each of the $\dot{\omega}$ values, and forwards integrated the values to get each respective $\omega$ value. After a satisfactory number of omega values were found, the results were plotted against time, and the period of oscillation in the resulting $\omega_x$ plot was found.

```python
import numpy as np
import matplotlib.pyplot as plt
```

The 3 equations of motion come that were used in computational assignment 8 are

(1) $I_1 \dot{\omega_1} = -\omega_2\omega_3(I_3-I_2)$

(2) $I_2 \dot{\omega_2} = -\omega_1\omega_3(I_1-I_3)$

(3) $I_3 \dot{\omega_3} = -\omega_1\omega_2(I_2-I_1)$

In order to use these, the diagonal terms, $I^{11}, I^{22},$ and $I^{33}$, must be found from the moment of inertia tensor. For this example, these diagonal terms were found to be $\frac{1}{12}Mb^2$, $\frac{1}{12}Ma^2$, and $\frac{1}{12}M(a^2 + b^2)$, respectively.

To show how the above values were created, a sample calculation will be shown here. 

Setting $dm = \frac{M}{ab}dudv$,

$I^{11} = 4\int_{0}^{\frac{a}{2}} \int_{0}^{\frac{b }{2}}\frac{M}{ab}dxdy (y^2 + z^2)$

since there is no z component in this lamina, the z term goes to 0.

This integral becomes $\frac{M}{ab} \frac{2b}{2} \frac{2a^3}{8}\frac{1}{3} = \frac{1}{12}Ma^2$. 

Similar processes were conducted for the other diagonal elements. Note that $I^{11}, I^{22},$ and $I^{33}$ are denoted as $I_1, I_2,$ and $I_3$.

```python
#Constants
M = 1 #in kg
a = 1 #in m
b = 2 #in m
#Initial conditions
w_x, w_y, w_z = 10, 0, 1 #in Hz
```

```python
def evo(w_x,w_y,w_z,M,a,b):
    #Finding the diagonal moment of inertia values using i.c.
    I1 = (1/12) * M * b**2
    I2 = (1/12) * M * a**2
    I3 = (1/12) * M * (a**2 + b**2)
    #Timestep
    ts = 1e-4
    t = 0
    #Lists to hold omega values
    wx, wy, wz = [], [], []
    
    while t <= 5:
        #Applying eom
        wx_d = -w_y * w_z * (I3 - I2) / I1
        wy_d = -w_x * w_z * (I1 - I3) / I2
        wz_d = -w_x * w_y * (I2 - I1) / I3
        #Forward integrating the omega values
        w_x = wx_d * ts + w_x
        w_y = wy_d * ts + w_y
        w_z = wz_d * ts + w_z
        #Adding the values to their respective lists
        wx.append(w_x)
        wy.append(w_y)
        wz.append(w_z)
        #Incrementing timesteps
        t += ts
        
    return wx,wy,wz
```

```python
t = np.linspace(0,5,50000)
#Calling the above function
w1, w2, w3 = evo(w_x,w_y,w_z,M,a,b)
```

```python
plt.figure(figsize = (12,5))
plt.plot(t,w1,label = "$\\omega_x$")
plt.plot(t,w2,label = "$\\omega_y$")
plt.plot(t,w3,label = "$\\omega_z$")
plt.title("Components of angular velocity as a function of time")
plt.xlabel("Time (s)")
plt.ylabel("$\\omega$ value (Hz)")
plt.grid()
plt.legend()
```

Now looking at strictly $\omega_x$

```python
plt.figure(figsize = (12,5))
plt.plot(t,w1,label = "$\\omega_x$")
plt.title("x-component of angular velocity as a function of time")
plt.xlabel("Time (s)")
plt.ylabel("$\\omega_x$ value (Hz)")
plt.grid()
plt.legend()
```

To more closely observe the time it takes for one full oscillation in the above plot, the plot's xrange is minimized.

```python
plt.figure(figsize = (12,5))
plt.plot(t,w1,label = "$\\omega_x$")
plt.title("x-component of angular velocity as a function of time")
plt.xlabel("Time (s)")
plt.xlim(0,2)
plt.ylabel("$\\omega_x$ value (Hz)")
plt.grid()
plt.legend()
```

It can be seen from this plot that the time is takes to go from 10Hz to -10Hz back to 10Hz, using the middle of the 10Hz plateau as the starting and ending point, is approximately 1.75 seconds.
