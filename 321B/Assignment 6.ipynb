```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Functions representing the eigenfrequencies
#+'ve case'
def w1(mu,g,l):
    delt = g/l
    return delt * (1 + (1/mu) + np.sqrt((1/mu) * (1 + (1/mu))))
#-'ve case'
def w2(mu,g,l):
    delt = g/l
    return delt * (1 + (1/mu) - np.sqrt((1/mu) * (1 + (1/mu))))
```

```python
g = 9.81 #in m/s^2
l = 1 #in m
#Arrays for the ratio of masses (in kg)
mu = np.linspace(1,100,1000)
```

```python
#Eigenfreq. data
w_1 = []
w_2 = []
#Finding each freq. for each mass ratio change
for i in mu:
    w_1.append(w1(i,g,l))
    w_2.append(w2(i,g,l))
```

```python
#plotting
plt.figure(figsize = (12,6))
plt.plot(mu,w_1,label = "$w_1$")
plt.plot(mu,w_2,label = "$w_2$")
plt.legend()
plt.grid()
plt.xlabel("$\\frac{m_1}{m_2}$")
plt.ylabel("$w^2$")
plt.title("Eigenfrequencies as a function of $\mu$")
```
